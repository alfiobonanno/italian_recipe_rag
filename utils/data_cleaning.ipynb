```python
import pandas as pd
```

```python
df = pd.read_csv('italian_recipes.csv')
df.head()
```

```python
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
import ast
df['Ingredienti'] = df['Ingredienti'].apply(ast.literal_eval)
```

```python
df["Ingredienti_formattati"] = df["Ingredienti"].apply(
    lambda ing: "\n".join([f"- {name}: {qty}" for name, qty in ing])
)
```

```python
df
```

```python
df['Ingredienti_formattati'].iloc[0]
```

```python
df["embed_text"] = (df["Nome"] + "\n\n" +"Ingredienti:\n" + df["Ingredienti_formattati"] + "\n\n" + "Procedimento:\n" + df["Steps"])
```

```python
df
```

```python
df['embed_text'].iloc[0]
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv('italian_recipes_cleaned.csv', index=False)
```

```python
from tqdm.auto import tqdm
import ollama

tqdm.pandas()  

model_name = "snowflake-arctic-embed2:568m"

df["embeddings"] = df["embed_text"].progress_apply(
    lambda x: ollama.embed(model=model_name, input=x).get("embeddings") or ollama.embed(model=model_name, input=x).get("embedding")
)
```

```python
df
```

```python
df['embeddings'] = df['embeddings'].apply(
    lambda vec: vec[0] if isinstance(vec[0], list) else vec
)
df
```
